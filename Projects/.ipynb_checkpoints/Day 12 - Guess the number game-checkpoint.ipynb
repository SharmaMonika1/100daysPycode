```python
import random
from art import logo_guessno

easy_turns = 10
hard_turns = 5

def set_difficulty():
    level = input("Choose a difficulty. Type 'easy' or 'hard': ")
    if level == "easy":
        return easy_turns
    else:
        return hard_turns
def compare_values(actual,guess,turns):
    if actual > guess:
        print("Too low.")
        return turns - 1
    elif actual < guess:
        print("Too high.")
        return turns - 1
    else:
        print(f"You got it! The answer is {actual}.")

def game():
    print(logo_guessno)
    print("Welcome to the Number Guessing Game!")
    print("I'm thinking of a number between 1 and 100.")
    actual_number = random.randint(0,101) 
    print(actual_number)     
    attempts = int(set_difficulty())
    guess_number = 0
    while guess_number != actual_number:
        print(f"You have {attempts} attempts remaining to guess the number.")
        guess_number = int(input("Make a guess:"))
        attempts = compare_values(actual_number,guess_number,attempts)
        if attempts ==0:
            print("You've run out of guesses, you lose.")
            return
        elif guess_number != actual_number:
            print("Guess again.")
game()
    
```
